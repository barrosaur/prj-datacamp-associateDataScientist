```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv", index_col=0)
# make it into a pandas data frame
netflix = pd.DataFrame(netflix_df)
```

```python
# access column duration and release_year and filter out the ones in 1990
netflix_1990 = netflix[np.logical_and(netflix["release_year"] >= 1990, netflix["release_year"] <= 1999)]
movies = netflix_1990[netflix_1990["type"] == "Movie"]
duration = movies[["duration"]].mode()
duration = duration[duration["duration"] != 0].iloc[0, 0]
```

```python
# short action movies
action_movies = netflix[np.logical_and(netflix["type"] == "Movie", netflix["genre"] == "Action")]

# release year
action_movie_1990 = action_movies[np.logical_and(action_movies["release_year"] >= 1990, action_movies["release_year"] <= 1999)]

# duration
action_movie_less_90 = action_movie_1990[action_movie_1990["duration"] < 90]

# short movie count
short_movie_count = action_movie_less_90.shape[0]

print(short_movie_count)
```
